### 1.2 数据预处理

模型解释：考虑到所选模型为logestic regression，是一个回归模型，因‘Year’对应的数据意为年份，在数学上与预测结果没有一定的相关性，故在回归时**不使用**该列。

```python
#除去'Year'列
cal_data=data.drop('Year',axis=1)
```

```python
#将Direction列转换为对应的0-1数据
cal_data[cal_data['Direction']=='Up']=1
cal_data[cal_data['Direction']=='Down']=0
cal_data['Direction']=cal_data['Direction'].astype('float64')
cal_data
```

## 2.模型训练与评估

```python
#拆分训练集与数据集
from sklearn.model_selection import train_test_split
Y=cal_data['Direction']
X=cal_data.drop('Direction',axis=1)
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1,random_state=1)
```

```python
#训练数据
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, Y_train)
```

```python
## 查看其对应的参数
print('the weight of Logistic Regression:',clf.coef_)
print('the intercept(w0) of Logistic Regression:',clf.intercept_)
```

```python
## 在训练集和测试集上分布利用训练好的模型进行预测
train_predict = clf.predict(X_train)
test_predict = clf.predict(X_test)
```

```python
#计算模型预测的精确度
from sklearn import metrics
print('The accuracy of the Logistic Regression for Traning Set is: %d%%' % (metrics.accuracy_score(Y_train,train_predict)*100))
print('The accuracy of the Logistic Regression for Test Set is: %d%%' % (metrics.accuracy_score(Y_test,test_predict)*100))
```

在训练集和测试集上，模型的精确度都为100%，模型拟合很完美，期待在新的数据上也有很好的表现效果！
