### 导包

```python
from statsmodels.tsa.stattools import acf,pacf
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import numpy as np
import matplotlib.pyplot as plt
```

### 生成白噪声序列

```python
def white_noise_series(e_num,e_len):
    '''生成服从标准正太分布的白噪声序列
    e_num为白噪声序列数，e_len为单个序列长度'''
    e=[]
    for i in range(e_num):
        e.append(np.random.randn(e_len))
        
    return e
```

```python
#测试：生成4个长度为50的白噪声序列
e=white_noise_series(4,50)
print(e)
```

```python
#绘制图像
plt.plot(e[0])
plt.show()
```

### 作业4

> 说明：针对每个MA模型，给出三组白噪声序列来绘制其ACF与PACF图像，每个白噪声序列长度为50

```python
e=white_noise_series(3,50)
```

(1)

```python
fig, ax = plt.subplots(3,2,figsize=(18,12))
for i in range(3):
    X_1=[]
    for j in range(len(e[i])-1):
        X_1.append(e[i][j]-2*e[i][j+1])
    
    plot_acf(np.array(X_1), ax=ax[i,0])
    plot_pacf(np.array(X_1), ax=ax[i,1])
```

(2) 

```python
fig, ax = plt.subplots(3,2,figsize=(18,12))
for i in range(3):
    X_2=[]
    for j in range(len(e[i])-1):
        X_2.append(e[i][j]-(1/2)*e[i][j+1])
    
    plot_acf(np.array(X_2), ax=ax[i,0])
    plot_pacf(np.array(X_2), ax=ax[i,1])
```

(3)

```python
fig, ax = plt.subplots(3,2,figsize=(18,12))
for i in range(3):
    X_3=[]
    for j in range(len(e[i])-2):
        X_3.append(e[i][j]-(4/5)*e[i][j+1]+(16/25)*e[i][j+2])
    
    plot_acf(np.array(X_3), ax=ax[i,0])
    plot_pacf(np.array(X_3), ax=ax[i,1])
```

(4)

```python
fig, ax = plt.subplots(3,2,figsize=(18,12))
for i in range(3):
    X_4=[]
    for j in range(len(e[i])-2):
        X_4.append(e[i][j]-(5/4)*e[i][j+1]+(25/16)*e[i][j+2])
    
    plot_acf(np.array(X_4), ax=ax[i,0])
    plot_pacf(np.array(X_4), ax=ax[i,1])
```

### 作业5


```python
e=white_noise_series(3,50)

fig, ax = plt.subplots(3,2,figsize=(18,12))
for i in range(3):
    X=[]
    X.append(np.random.rand())
    
    for j in range(len(e[i])-1):
        X.append(0.5*X[j]+e[i][j]-0.8*e[i][j+1])
    
    plot_acf(np.array(X), ax=ax[i,0])
    plot_pacf(np.array(X), ax=ax[i,1])
```

可以看出：

ARMA(1,1)模型的ACF与PACF始终有非零取值，不会趋近或恒等于0，即**ARMA(1,1)模型的ACF与PACF均具有拖尾性**
